```python
import tkinter as tk
from tkinter import filedialog,font,colorchooser,messagebox
from tkinter.ttk import *
from PIL import Image,ImageTk
import os
import tempfile
#functions
def print_file(event=None):
    file=tempfile.mktemp('.txt')
    open(file,'w').write(text.get(1.0,tk.END))
    os.startfile(file,'print')
def theme(bg_color,fg_color):
    text.config(bg=bg_color,fg=fg_color)
def choose_color():
    color_code = colorchooser.askcolor(title="Choose Color")
    if color_code[1]:  # Check if a color was chosen
        text.config(fg=color_code[1])
file_path=""           
def newfile(event=None):
    text.delete(1.0,tk.END)   
def open_file(event=None):
    global file_path
    file_path=filedialog.askopenfilename(filetypes=[("Text Files","txt"),('All Files','*.*')])
    if file_path!='':
        data=open(file_path,"r")
        text.insert(0.0,data.read())
        root.title(os.path.basename(file_path))
def save_file(event=None):
    if file_path:
        #content=text.get(0.0,tk.END)
        #file_path.write(content)
        #file_path.close()
   # else:
        content=text.get(0.0,tk.END)
        file=open(file_path,'w')
        file.write(content)
def saveas_file(event=None):
    if file_path=='':
        save_path=filedialog.asksaveasfilename(defaultextension=".txt",filetype=[("Text Files","txt"),('All Files','*.*')])
        content=text.get(0.0,tk.END)
        save_path.write(content)
        save_path.close()
        if file_path!='':
            os.remove(file_path)
def exit(event=None):
    if text.edit_modified():
        result=messagebox.askyesnocancel('Warning','Do you want to save the file?')
        if result is True:
            if file_path=='':
                content=text.get(1.0,tk.END)
                file=open(file_path)
                file.write(content)
                root.destroy()
            else:
                content=text.get(0.0,tk.END)
                save_path=filedialog.asksaveasfilename(defaultextension=".txt",filetype=[("Text Files","txt"),('All Files','*.*')])
                save_path.write(content)
                save_path.close()
                root.destroy()
        elif result is False:
            root.destroy()
        else:
            pass        
fontsize=10
fontstyle="Arial"
def font_style(event):
    global fontstyle
    font_style=fontfamily_variable.get()
    text.configure(font=(fontstyle,fontsize))
def font_size(event):
    global fontsize
    fontsize=int(size_variable.get())
    text.configure(font=(fontstyle,fontsize)) 
def bold_text():
    current_tags=text.tag_names("sel.first")
    text_pro=font.Font(text,text.cget("font"))
    text_pro.configure(weight='bold')
    text.tag_configure("bold",font=text_pro) 
    if "bold" in current_tags:
        text.tag_remove("bold","sel.first","sel.last")
    else:
        text.tag_add("bold","sel.first","sel.last")   
def italic_text():
    current_tags=text.tag_names("sel.first")
    text_pro=font.Font(text,text.cget("font"))
    text_pro.configure(slant='italic')
    text.tag_configure("italic",font=text_pro) 
    if "italic" in current_tags:
        text.tag_remove("italic","sel.first","sel.last")
    else:
        text.tag_add("italic","sel.first","sel.last")
def underline_text():
    current_tags=text.tag_names("sel.first")
    text_pro=font.Font(text,text.cget("font"))
    text.tag_configure("underline",underline=True) 
    if "underline" in current_tags:
        text.tag_remove("underline","sel.first","sel.last")
    else:
        text.tag_add("underline","sel.first","sel.last")
def align_right():
    data=text.get(0.0,tk.END)
    text.tag_config('right',justify="right")
    text.delete(0.0,tk.END)
    text.insert(tk.INSERT,data,'right')
def align_left():
    data=text.get(0.0,tk.END)
    text.tag_config('left',justify="left")
    text.delete(0.0,tk.END)
    text.insert(tk.INSERT,data,'left')
def align_center():
    data=text.get(0.0,tk.END)
    text.tag_config('center',justify="center")
    text.delete(0.0,tk.END)
    text.insert(tk.INSERT,data,'center')  
def find(event):
    def find_words():
        start_pos='1.0'
        word=entryfield.get()
        while True:
            start_pos=text.search(word,start_pos,stopindex="end")
            if not start_pos:
                break
            end_pos=f'{start_pos}+{len(word)}c'
            text.tag_add('match',start_pos,end_pos)
            text.tag_config('match',foreground='blue',background='pink')
            start_pos=end_pos
    def replace():
        word=entryfield.get()
        replaceword=replaceentryfield.get()
        content=text.get(1.0,tk.END)
        new=content.replace(word,replaceword)
        text.insert(1.0,new)
        
    #GUI
    root1=tk.Toplevel()
    root1.title("Find")
    root1.geometry('450x250+500+200')
    root1.resizable(0,0)
    labelFrame=LabelFrame(root1,text="Find/Replace")
    labelFrame.pack(pady=10)
    findlabel=Label(labelFrame,text="Find")
    findlabel.grid(row=0,column=0,padx=5,pady=5)
    entryfield=Entry(labelFrame)
    entryfield.grid(row=0,column=1,padx=5,pady=5)
    replacelabel=Label(labelFrame,text="Replace")
    replacelabel.grid(row=1,column=0,padx=5,pady=5)
    replaceentryfield=Entry(labelFrame)
    replaceentryfield.grid(row=1,column=1,padx=5,pady=5)
    findbutton=Button(labelFrame,text='FIND',command=find_words)
    findbutton.grid(row=2,column=0,padx=5,pady=5)
    replacebutton=Button(labelFrame,text='REPLACE',command=replace)
    replacebutton.grid(row=2,column=1,padx=5,pady=5)
    root1.mainloop()
def toolfunc():
    if show_toolbar.get()==False:
        tool_bar.pack_forget()
    if show_toolbar.get()==True:
        text.pack_forget()
        tool_bar.pack(fill='x')  
        text.pack(fill='both',expand=1)     
root=tk.Tk()
root.title("Text Editor")
root.geometry("1000x500+10+10")
menu=tk.Menu(root)
root.config(menu=menu)
file_menu=tk.Menu(menu)
edit_menu=tk.Menu(menu)
view_menu=tk.Menu(menu)
theme_menu=tk.Menu(menu)
insert_menu=tk.Menu(menu)
# Create a text widget
text=tk.Text(root,wrap=tk.WORD)
scroll=Scrollbar(root)
scroll.pack(side="right",fill="y")
text=tk.Text(root,yscrollcommand=scroll.set)
scroll.config(command=text.yview)
text.pack(expand=True,fill="both")
#File Menu
menu.add_cascade(label="File",menu=file_menu)
file_menu.add_command(label="New",accelerator="Ctrl+N",command=newfile)
file_menu.add_command(label="Open",accelerator="Ctrl+O",command=open_file)
file_menu.add_command(label="Save",accelerator="Ctrl+S",command=save_file)
file_menu.add_command(label="Save As",accelerator="Ctrl+Alt+S",command=saveas_file)
file_menu.add_command(label="Print",accelerator="Ctrl+P",command=print_file)
file_menu.add_separator()
file_menu.add_command(label="Exit",accelerator="Ctrl+Q",command=exit)
#Edit Menu
menu.add_cascade(label="Edit",menu=edit_menu)
edit_menu.add_command(label="Cut",accelerator="Ctrl+X",command=lambda:text.event_generate('<Control x>'))
edit_menu.add_command(label="Copy",accelerator="Ctrl+C",command=lambda:text.event_generate('<Control c>'))
edit_menu.add_command(label="Paste",accelerator="Ctrl+V",command=lambda:text.event_generate('<Control v>'))
edit_menu.add_command(label="Select All",accelerator="Ctrl+A",command=lambda:text.event_generate('<Control a>'))
edit_menu.add_command(label="Find",accelerator="Ctrl+F",command=find)
#View Menu
menu.add_cascade(label="View",menu=view_menu)
show_toolbar=tk.BooleanVar()
view_menu.add_checkbutton(label="Tool Bar",variable=show_toolbar,onvalue=True,offvalue=False,command=toolfunc)
show_toolbar.set(True)
#Theme Menu
menu.add_cascade(label="Themes",menu=theme_menu)
theme_choice=tk.StringVar()
theme_menu.add_radiobutton(label="Light Default",variable=theme_choice,command=lambda:theme('white','black'))
theme_menu.add_radiobutton(label="Dark",variable=theme_choice,compound="left",command=lambda:theme('gray','white'))
theme_menu.add_radiobutton(label="Pink",variable=theme_choice,compound="left",command=lambda:theme('pink','blue'))
theme_menu.add_radiobutton(label="Monokai",variable=theme_choice,compound="left",command=lambda:theme('green','white'))
#Tool bar
tool_bar=tk.Label(root)
tool_bar.pack(fill='x',side='top')
font_family=font.families()
fontfamily_variable=tk.StringVar()
fontfamily=Combobox(tool_bar,width=30,values=font_family,textvariable=fontfamily_variable)
fontfamily.current(font_family.index("Arial"))
fontfamily.grid(row=0,column=0,padx=5)
size_variable=tk.IntVar()
fontsize=Combobox(tool_bar,width=30,textvariable=size_variable,values=tuple(range(8,72)))
fontsize.current(2)
fontsize.grid(row=0,column=1,padx=5)
fontfamily.bind("<<ComboboxSelected>>",font_style)
fontsize.bind("<<ComboboxSelected>>",font_size)


bold_image=ImageTk.PhotoImage(file="C:/Users/shree/Downloads/bold1.png")
button1=Button(tool_bar,image=bold_image,command=bold_text)
button1.grid(row=0,column=2,padx=5)

italicimage=ImageTk.PhotoImage(file="C:/Users/shree/Downloads/italic.png")
button2=Button(tool_bar,image=italicimage,command=italic_text)
button2.grid(row=0,column=3,padx=5)

colorimage=tk.PhotoImage(file="C:/Users/shree/Downloads/color.png")
button3=Button(tool_bar,image=colorimage,command=choose_color)
button3.grid(row=0,column=4,padx=5)

underlineimage=ImageTk.PhotoImage(file="C:/Users/shree/Downloads/underline.png")
button4=Button(tool_bar,image=underlineimage,command=underline_text)
button4.grid(row=0,column=5,padx=5)

leftimage=ImageTk.PhotoImage(file="C:/Users/shree/Downloads/left.png")
button5=Button(tool_bar,image=leftimage,command=align_left)
button5.grid(row=0,column=6,padx=5)

centerimage=ImageTk.PhotoImage(file="C:/Users/shree/Downloads/center.png")
button6=Button(tool_bar,image=centerimage,command=align_center)
button6.grid(row=0,column=7,padx=5)

rightimage=ImageTk.PhotoImage(file="C:/Users/shree/Downloads/right.png")
button7=Button(tool_bar,image=rightimage,command=align_right)
button7.grid(row=0,column=8,padx=5)

root.bind('<Control-o>',open_file)
root.bind('<Control-n>',newfile)
root.bind('<Control-s>',save_file)
root.bind('<Control-Alt-s>',saveas_file)
root.bind('<Control-p>',print_file)
root.bind('<Control-q>',exit)
root.bind('<Control-f>',find)


root.mainloop()
    
```
